# Area Plots With Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

# Create a quick DataFrame
df = pd.DataFrame(randn(100, 4), columns=["Monday", "Tuesday", "Wednesday", "Thursday"])
df
```

## Area plot

```python
# Area plot
# Positive and negative values 
# 1st method: stacked=False
df.plot(kind='area', stacked=False)
```

```python
# Area plot
# Positive and negative values 
# 2nd mathod: convert values to absolute values
df.abs().plot(kind='area')
```

```python
# Shading (0 to 1, 0 light alpha, 1 no alpha)
df.abs().plot(kind='area', alpha=0.3)
```

```python
# Shading
df.abs().plot(kind='area', alpha=0.8)
```

```python
# Title
df.abs().plot(kind='area', alpha=0.5, title="My Awesome Area Plot")
```

```python
# Legend (on or off, True by default)
df.abs().plot(kind='area', alpha=0.5, title="My Awesome Area Plot", legend=False)
```

```python
# Plot only one column
df.Monday.abs().plot(kind='area', alpha=0.5, title="My Awesome Area Plot")
```

```python
# Plot two columns
df[["Monday", "Thursday"]].abs().plot(kind='area', alpha=0.5, title="My Awesome Area Plot")
```
