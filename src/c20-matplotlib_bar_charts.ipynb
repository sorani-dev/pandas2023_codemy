# Histograms With Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

# Create a quick DataFrame
df = pd.DataFrame(randn(100, 4), columns=["Monday", "Tuesday", "Wednesday", "Thursday"])
df
```

## Bar plot

```python
# Bar plot
df.plot(kind='bar', figsize=(10, 10))
```

```python
# Bar plot (positive values)
df.abs().plot(kind='bar')
```

## Bar plot stacked

```python
# Bar plot stacked
df.abs().plot(kind='bar', stacked=True, figsize=(10, 10))
```

## Shading

```python
# shading
df.abs().plot(kind='bar', stacked=True, alpha=0.4)
```

## Title

```python
# Title
df.abs().plot(kind='bar', stacked=True, title="My Awesome Bar Chart")
```

## No Legend (legend=False)

```python
# No Legend (legend=False)
df.abs().plot(kind='bar', stacked=True, legend=False, figsize=(20, 20))
```

```python
# No Legend (legend=False)
df.abs().plot(kind='bar', stacked=True, legend=False, figsize=(10, 10))
```

## Second way

```python
# Second way
df.abs().plot.bar(figsize=(12, 12))
```

```python
## Second way (with function bar()) stacked
```

```python
# Second way (with function bar()) stacked
df.abs().plot.bar(stacked=True, figsize=(12, 12))
```

## Second way (with function bar()) shading

```python
# Second way (with function bar()) shading
df.abs().plot.bar(stacked=True, figsize=(12, 12))
```

## Second way (with function bar()) title

```python
# Second way (with function bar()) title
df.abs().plot.bar(stacked=True, title="My Awesome Bar Chart", figsize=(12, 12))
```

## Second way (with function bar()) no legend

```python
# Second way (with function bar()) no legend
df.abs().plot.bar(stacked=True, legend=False, figsize=(12, 12))
```

## Second way (with function bar()) Grid lines

```python
# Second way (with function bar()) Grid lines
df.abs().plot.bar(stacked=True, grid=True, figsize=(12, 12))
```
