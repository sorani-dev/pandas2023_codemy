# Scatterplots With Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

# Create a quick DataFrame
df = pd.DataFrame(randn(500, 4), columns=["Monday", "Tuesday", "Wednesday", "Thursday"])
df
```

## Scatterplot

```python
# Scatterplot
df.plot(kind='scatter', x='Monday', y='Tuesday')
```

```python
# Scatterplot
df.plot(kind='scatter', x='Monday', y='Tuesday', c='Wednesday')
```

## Colors: 
(https://matplotlib.org/stable/api/colors_api.html)
(https://matplotlib.org/stable/tutorials/colors/index.html)
(https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
# Colors (https://matplotlib.org/stable/api/colors_api.html)
# https://matplotlib.org/stable/tutorials/colors/index.html
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
df.plot(kind='scatter', x='Monday', y='Tuesday', c='Wednesday', cmap='magma')
```

```python
# Colors (https://matplotlib.org/stable/api/colors_api.html)
# https://matplotlib.org/stable/tutorials/colors/index.html
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
df.plot(kind='scatter', x='Monday', y='Tuesday', c='Wednesday', colormap='inferno')
```

```python
# Colors (https://matplotlib.org/stable/api/colors_api.html)
# https://matplotlib.org/stable/tutorials/colors/index.html
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
df.plot(kind='scatter', x='Monday', y='Tuesday', c='Wednesday', colormap='Greens')
df.plot(kind='scatter', x='Monday', y='Tuesday', c='Wednesday', colormap='summer')
df.plot(kind='scatter', x='Monday', y='Tuesday', c='Wednesday', colormap='PuOr')
```

## By Size

```python
# By Size
df.plot(kind='scatter', x='Monday', y='Tuesday', s=df['Wednesday'] * 100)
```

## By Size easier to read with Alpha channel

```python
# By Size easier to read with Alpha channel
df.plot(kind='scatter', x='Monday', y='Tuesday', s=df['Wednesday'] * 100, alpha = 0.4)
```

```python
## Title
```

```python
# Title
df.plot(kind='scatter', x='Monday', y='Tuesday', s=df['Wednesday'] * 100, alpha = 0.4, title='My Awesome ScatterPlot')
```
