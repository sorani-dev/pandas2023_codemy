# Hex Bin Plots With Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

# Create a quick DataFrame
df = pd.DataFrame(randn(500, 4), columns=[
                  "Monday", "Tuesday", "Wednesday", "Thursday"])
df
```

## Hexplots

```python
# Hexplots
df.plot(kind='hexbin', x='Monday', y='Tuesday')
```

## Hexplots gridsize
The lower the gridsize, the bigger the bins (points)

```python
# Hexplots gridsize
df.plot(kind='hexbin', x='Monday', y='Tuesday', gridsize=10)
```

```python
# Hexplots gridsize
df.plot(kind='hexbin', x='Monday', y='Tuesday', gridsize=25)
```

## Hexplots color

```python
# Hexplots color
df.plot(kind='hexbin', x='Monday', y='Tuesday', gridsize=25, colormap='magma')
```

## Other method

```python
# Other method
df.plot.hexbin(x='Monday', y='Tuesday')
df.plot.hexbin(x='Monday', y='Tuesday', gridsize=10)
help(df.plot.hexbin)
```

### with title

```python
# With title
df.plot.hexbin(x='Monday', y='Tuesday', gridsize=25,
               title='My Awesome Hex Plot')
```

### With title and grid

```python
# With title and grid
df.plot.hexbin(x='Monday', y='Tuesday', gridsize=25,
               title='My Awesome Hex Plot', grid=True)
```
