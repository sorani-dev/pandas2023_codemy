# Histograms With Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

# Create a quick DataFrame
df = pd.DataFrame(randn(100, 4), columns=["Monday", "Tuesday", "Wednesday", "Thursday"])
df
```

## Histogram

```python
# Histogram
df.Wednesday.hist()
```

## Histogram with smaller Bins, 50, 20, 100

```python
# Histogram with smaller Bins, 50, 20
df.Wednesday.hist(bins=50)
```

```python
df.Wednesday.hist(bins=20)
```

```python
df.Wednesday.hist(bins=100)
```

## Another way without grid bars

```python
# Another way (plot) without grid bars
df.Wednesday.plot(kind='hist')
```

```python
# Another way (plot) without grid bars
df.Wednesday.plot(kind='hist', bins=20)
```

```python
# Another way without grid bars
df.Wednesday.plot(kind='hist', bins=30, grid=False)
```

```python
# Another way without grid bars and Legend
df.Wednesday.plot(kind='hist', bins=30, grid=False, legend=True)
```
