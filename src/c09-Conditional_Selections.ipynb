# Conditional Selections

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/dog_data.csv')
df
```

## Conditional Selection
Operators: `>, <, ==, >=, <=,` etc

Returns Booleans

```python
# Conditional Selection
# > < == >= <= etc
# Returns Booleans
df == "BROWN"
```

## Return DataFrame with data

```python
# Return DataFrame with data
df[df == "BROWN"]
```

## Run on a column 

```python
# Run on a column 
df[df == "BROWN"]["Color"]
```

## Return mutiple columns

```python
df[df == "BROWN"][["Color", "Breed"]]
```

## Return mutiple columns

```python
df[df["OwnerZip"] > 70000][["Color", "Breed", "OwnerZip"]]
```
