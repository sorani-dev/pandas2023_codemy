# Sorting And Ordering Data

```python
import pandas as pd
import numpy as np
# Create Data
stuff = {
    'Corporation':['Apple', 'Google', 'Meta', 'Apple', 'Google','Meta'],
    'Employees':['John', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    'Salary':[200, 220, 190, 130, 120, 150]}
# Create Dataframe
df = pd.DataFrame(stuff)
df
```

## Sort Salary from least to highest (default ascending order)

```python
# Sort Salary from least to highest (default ascending order)
df.sort_values(by="Salary"), \
df.sort_values("Salary")
```

## Sort Salary from highest to least

```python
# Sort Salary from highest to least
df.sort_values(by="Salary", ascending=False)
```

## Not permanent unless inplace=True

```python
# Not permanent unless inplace=True
df
```

## With inplace=True
Salary

```python
# Inplace
df.sort_values("Salary", inplace=True)
df
```

## With inplace=True
Employees

```python
# Inplace
df.sort_values("Employees", inplace=True)
df
```

```python
df.sort_values("Employees", inplace=True, ascending=False)
df
```
