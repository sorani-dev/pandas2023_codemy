# Add new columns to DataFrame

```python
import pandas as pd
import numpy as np
# import csv file
df = pd.read_csv('data/dog_data_short.csv')
df
```

Add Column from a List

```python
# Add Column from a List
gender = ["Male", "Female","Male","Male", "Female"]
df["Gender"] = gender
df
```

Add Default Values

```python
# Add Default Values
df["Alive"] = [True] * len(df)
df
```

Add NaN values

```python
# Add NaN values
df["Show Dog"] = [np.nan] * len(df)
df
```

Add column with **.insert()** - Allows position

```python
# Add column with .insert() - Allows position
df.insert(1, "Adopted", [True] * len(df), True)
df.insert(loc=1, column="Adopted 2", value=[True] * len(df), allow_duplicates=True)
df
```

Add column with **.assign()** - Create new DataFrame

```python
# Add column with .assign() - Create new DataFrame
df2 = df.assign(Horse=[False] * len(df))
df2
```
