# Density and KDE Plots With Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

# Create a quick DataFrame
df = pd.DataFrame(randn(500, 4), columns=[
                  "Monday", "Tuesday", "Wednesday", "Thursday"])
df
```

## Density and kernel density plots with scipy

```python
# Density and kernel density plots with scipy
df.plot(kind='kde')
```

### KDE second way to learn shift tab

```python
# KDE second way to learn shift tab
df.plot.kde()
```

```python
help(df.plot.kde)
```

## Density plot

```python
df.plot.density()
```

```python
help(df.plot.density)
```

## Other method

```python
# Other method
df.plot(kind='density')
```

## KDE 
### Change alpha channel, width

```python
# KDE: Change alpha channel, width
df.plot.density(alpha=.4, lw=4)
```

### Only one category

```python
# One category
df['Wednesday'].plot.density(alpha=.4, lw=4)
```

### Grid

```python
# Grid
df['Tuesday'].plot.kde(alpha=0.4, lw=4, grid=True)
```

### Title

```python
# Title
df['Tuesday'].plot.kde(alpha=0.4, lw=4, title='My Awesome KDE Plot')
```

```python
# Title
df['Tuesday'].plot.kde(alpha=0.4, lw=4, grid=True, title='My Awesome KDE Plot')
```
