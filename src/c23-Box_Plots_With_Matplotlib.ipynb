# Box Plots With Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

# Create a quick DataFrame
df = pd.DataFrame(randn(500, 4), columns=[
                  "Monday", "Tuesday", "Wednesday", "Thursday"])
df
```

## Boxplots

```python
# Boxplots
df.plot(kind='box')
```

```python
# Boxplots size
```

```python
# Boxplots size
df.plot(kind='box', figsize=(10, 10))
```

## Boxplot color

```python
# Boxplot color
df.plot(kind='box', cmap='jet')
df.plot(kind='box', colormap='jet')
```

## No legend

```python
# No legend
```

## Title

```python
# Title
df.plot(kind='box', title='My Awesome Box plot')
```

## Other way

```python
# Other way
df.plot.box()
df.plot.box(title="My Awesome Box Plot")
```

## Add Grid

```python
# Add Grid
df.plot.box(title="My Awesome Box Plot", grid=True)
```
